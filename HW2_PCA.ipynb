```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
from sklearn.decomposition import PCA
```

```python
data = pd.read_table('Nichols_Nicki.dat')
```

```python
data
data.describe().T.sort_values("std").head()
```

```python
variance = data.var()
variance
data1 = data.drop(['ias23','ias34'],axis=1)
data1
```

```python
matrix = data1.corr()
sorted_matrix = matrix.unstack().sort_values()
sorted_matrix
```

```python
data2 = data1.drop(['ias13','ias22','ias32'], axis=1)
data2.corr()
```

```python
np_data = np.cov(data2.T)
```

```python
eigen_vals, eigen_vecs = np.linalg.eig(np_data)
eigen_vals
```

```python
pca_n = PCA(n_components=None)
np_data_pca = pca_n.fit_transform(data2)
pca_n.explained_variance_ratio_
```

```python
pca_5 = PCA(n_components=5)
np_data_pca5 = pca_5.fit_transform(data2)
pca_5.explained_variance_ratio_
```
